```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# ---------------------------------------------------------
# 1. Data Simulation: Generating Synthetic Satellite Imagery
# ---------------------------------------------------------
# In a real scenario, you would use Rasterio to load .tif files.
# Here, we generate "Satellite Images" (RGB) and "Land Cover Masks".

def generate_synthetic_data(num_samples, img_size=128):
    """
    Generates synthetic satellite images and corresponding masks.
    Classes: 0=Background/Soil, 1=Forest (Green), 2=Urban (Gray), 3=Water (Blue)
    """
    images = []
    masks = []

    for _ in range(num_samples):
        # Create a blank image and mask
        img = np.ones((img_size, img_size, 3)) * 210 # Soil background
        mask = np.zeros((img_size, img_size, 1)) # Mask background (class 0)

        # Add random Forest patches (Class 1)
        for _ in range(np.random.randint(2, 5)):
            center = (np.random.randint(20, 108), np.random.randint(20, 108))
            radius = np.random.randint(10, 25)
            cv2.circle(img, center, radius, (34, 139, 34), -1) # Green
            cv2.circle(mask, center, radius, 1, -1) # Label 1

        # Add Urban structures (Class 2)
        for _ in range(np.random.randint(1, 4)):
            pt1 = (np.random.randint(0, 100), np.random.randint(0, 100))
            pt2 = (pt1[0] + np.random.randint(10, 30), pt1[1] + np.random.randint(10, 30))
            cv2.rectangle(img, pt1, pt2, (128, 128, 128), -1) # Gray
            cv2.rectangle(mask, pt1, pt2, 2, -1) # Label 2

        # Add Water bodies (Class 3)
        pts = np.random.randint(0, img_size, (5, 2))
        pts = pts.reshape((-1, 1, 2)).astype(np.int32)
        cv2.fillPoly(img, [pts], (65, 105, 225)) # Blue
        cv2.fillPoly(mask, [pts], 3) # Label 3

        # Add some noise to the image to simulate sensor noise
        noise = np.random.normal(0, 10, img.shape).astype(np.int16)
        img = np.clip(img.astype(np.int16) + noise, 0, 255).astype(np.uint8)

        images.append(img)
        masks.append(mask)

    return np.array(images), np.array(masks)

print("Generating synthetic satellite dataset...")
# Generate Training and Testing sets
X_train, y_train = generate_synthetic_data(100)
X_test, y_test = generate_synthetic_data(20)

# Normalize images
X_train = X_train / 255.0
X_test = X_test / 255.0

print(f"Train Data Shape: {X_train.shape}, Train Masks Shape: {y_train.shape}")

# ---------------------------------------------------------
# 2. Model: U-Net Architecture for Semantic Segmentation
# ---------------------------------------------------------
# U-Net is the standard CNN architecture for remote sensing/segmentation.

def unet_model(input_size=(128, 128, 3), num_classes=4):
    inputs = layers.Input(input_size)

    # --- Encoder (Feature Extraction) ---
    c1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    # --- Bottleneck ---
    c3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c3)

    # --- Decoder (Upsampling) ---
    u4 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c3)
    u4 = layers.concatenate([u4, c2]) # Skip Connection
    c4 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u4)
    c4 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c4)

    u5 = layers.Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c4)
    u5 = layers.concatenate([u5, c1]) # Skip Connection
    c5 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(u5)
    c5 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c5)

    # Output layer: 4 classes (softmax)
    outputs = layers.Conv2D(num_classes, (1, 1), activation='softmax')(c5)

    model = models.Model(inputs=[inputs], outputs=[outputs])
    return model

print("Building U-Net Model...")
model = unet_model()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# ---------------------------------------------------------
# 3. Training
# ---------------------------------------------------------
print("\nStarting Training...")
history = model.fit(X_train, y_train,
                    epochs=10,
                    batch_size=8,
                    validation_data=(X_test, y_test),
                    verbose=1)

# ---------------------------------------------------------
# 4. Change Detection Logic
# ---------------------------------------------------------
print("\nSimulating Change Detection (Deforestation Analysis)...")

# Select a sample image and predict the current land cover
sample_img = X_test[0]
pred_mask = model.predict(np.expand_dims(sample_img, axis=0))[0]
pred_mask_class = np.argmax(pred_mask, axis=-1)

# Simulate a "Past" scenario:
# We take the current mask and artificially remove some forest (Class 1) to simulate deforestation
past_mask = pred_mask_class.copy()
# Randomly clear a patch of forest to simulate urbanization or clearing
past_mask[40:80, 40:80] = 2 # Change area to Urban (Class 2) or Soil (Class 0)

# Calculate Change
# Detect areas that were Forest (1) in the past but are NOT Forest (1) now
change_map = np.zeros_like(past_mask)
# Where past was Forest (1) AND current is NOT Forest -> Detected Deforestation
change_map[(past_mask == 1) & (pred_mask_class != 1)] = 1

print(f"Deforestation Detected in {np.sum(change_map)} pixels.")

# ---------------------------------------------------------
# 5. Visualization & Evaluation
# ---------------------------------------------------------

# Visualizing Results
plt.figure(figsize=(15, 6))

# 1. Satellite Image
plt.subplot(1, 4, 1)
plt.imshow(sample_img)
plt.title("Satellite Image (Input)")

# 2. Predicted Land Cover
plt.subplot(1, 4, 2)
plt.imshow(pred_mask_class)
plt.title("Predicted Land Cover\n(1:Forest, 2:Urban)")

# 3. Simulated Past
plt.subplot(1, 4, 3)
plt.imshow(past_mask)
plt.title("Simulated Past Data\n(Showing Forest)")

# 4. Change Detection Map
plt.subplot(1, 4, 4)
plt.imshow(change_map, cmap='Reds')
plt.title("Detected Change\n(Red = Deforestation)")

plt.tight_layout()
plt.show()

# Evaluation Metrics (Flattening arrays for scikit-learn)
print("\nClassification Report for Land Cover Classification:")
print(classification_report(y_test.flatten(),
                            np.argmax(model.predict(X_test), axis=-1).flatten(),
                            target_names=['Soil', 'Forest', 'Urban', 'Water']))
```
